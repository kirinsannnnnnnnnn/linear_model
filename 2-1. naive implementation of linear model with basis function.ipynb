# 2-1. 多項式基底関数を用いた線形モデルによる回帰

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(1)
```

```python
X = np.linspace(0,2*np.pi,25)
t = np.sin(X) + np.random.normal(0,0.3,25)
plt.scatter(X, t)
plt.plot(X, np.sin(X))
```

## 線形モデルにより最尤解を求める

```python
def phi_multi(x, M):
    """
    M-1次多項式基底
    """
    return [np.power(x,i) for i in range(M)]
```

```python
M=4
PHI = np.array([phi_multi(x,M) for x in X])
```

```python
w = np.dot(
            np.linalg.inv(
                np.dot(PHI.T, PHI)
            ), 
            np.dot(PHI.T, t)
        )
```

```python
def f(w,x, M):
    return np.dot(w, phi_multi(x, M))
```

```python
xlist = np.linspace(0,2*np.pi, 100)
ylist = f(w,xlist, M)

plt.plot(xlist, ylist)
plt.scatter(X, t)
plt.plot(X, np.sin(X))
```
